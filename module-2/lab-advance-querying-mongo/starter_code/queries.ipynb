# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
list(collection_companies.find({'name':'Babelgum'},{'_id':0,'name':1}))
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
#Buscamos empresas por numero de empleados mayor a 5000, solo el nombre, sin el id, y ordenamos por numero 
#Se muestran el nombre y el numero de empleados para comprobar...
list(collection_companies.find({'number_of_employees':{'$gt':5000}},
                              {'_id':0,'name':1,'number_of_employees':1}).limit(20).sort('number_of_employees'))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
#Se muestran compañias fundadas entre 2000 y 2005 (inlucidos), se incluyen solo nombres y los años de fundacion... 
list(collection_companies.find({'founded_year':{'$gte':2000,'$lte':2005}},{'_id':0,'name':1,'founded_year':1}))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
#Se muestran las empresas fundadas de 2010 hacia atras, con un valor de activos superior a los 10 millones
list(collection_companies.find({'founded_year':{'$lte':2010}, 'ipo.valuation_amount':{'$gt':100000000}},
                               {'_id':0,'name':1,'ipo.valuation_amount':1,'founded_year':1}))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
list(collection_companies.find({'number_of_employees':{'$lt':1000},'founded_year':{'$lt':2005}},
    {'_id':0,'name':1,'number_of_employees':1,'founded_year':1}).sort('number_of_employees').limit(10))
```

### 6. All the companies that don't include the `partners` field.

```python

collection_companies.count_documents({'partners':{'$exists':None}})
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
list(collection_companies.find({'category_code':None},{'name':1,'_id':0}))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
list(collection_companies.find({'number_of_employees':{'$gte':100},'number_of_employees': {'$lt':1000}}, 
                               {'name':1, 'number_of_employees': 1, '_id':0}))
```

### 9. Order all the companies by their IPO price descendently.

```python
list(collection_companies.find({'ipo.valuation_amount':{'$gt':0}},
                               {'_id':0,'name':1, 'ipo.valuation_amount':1}).sort('ipo.valuation_amount',-1))
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
list(collection_companies.find({'number_of_employees': {'$gt': 0}},
                               {'_id': 0,'name': 1, 'number_of_employees': 1}).sort('number_of_employees', -1).limit(10))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
list(collection_companies.find({'founded_month':{'$gt':6}},
                               {'_id':0,'name':1,'founded_month':1,'founded_year':1}).limit(1000))
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
list(collection_companies.find({'$subtract':['deadpooled_year','founded_year']:{'$gt':3}},
                               {'_id':0,'founded_year':1,'deadpooled_year':1}))
```
