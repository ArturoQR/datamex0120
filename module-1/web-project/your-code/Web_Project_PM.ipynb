```python
#1.- Hacemos el request para el scraping de la lista de lugares que están registrados como pueblos mágicos...
#Aunque resultó ser un desorden, algunos pueblos indicados en revistas no aparecen en SecTUR
import requests

url = 'http://www.sectur.gob.mx/gobmx/pueblos-magicos/'
html = requests.get(url).content
html[0:600]
```

```python
#Comenzamos la limpieza con BeautifulSoup para dejar solo la lista...
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")
soup
```

```python
#Extraemos los lugares en la lista de pueblos...
tags = ['li']
text = [element.text for element in soup.find_all(tags)]
text
```

```python
#Retiramos los primeros 14 elementos ya que, estos correponden a indicaciones al inicio de la página...
text=text[13:]
text
```

```python
#Revisamos la longitud de la lista
len(text)
```

```python
#Separamos la lista en una lista de listas para poder crear un dataframe en los siguientes pasos...
new_list=[]
for each in text:
    new_list.append(each.split(', '))
    
new_list
```

```python
#Importamoe todo lo necesario para usar OpenStreetMap...
import os
import geopy
from geopy.geocoders import Nominatim
```

```python
#Hay que usar esta indicación de my-application, hacerlod e forma anónima ya no se permite...
nom=Nominatim(user_agent="my-application")
```

```python
#Se define el elemento Nominatim, el que permite conectarse a OpenStreetMap...
#Esta fue una prueba...
n=nom.geocode('Cuatro Ciénegas, Coahuila')
```

```python
#Hacemos un ciclo for para obtener el máximo posible de ubicaciones, algunos pueblos no son
#identificados correctamente o seguramente debe de indicarse con mayor precisión...
locs=[]
for pm in text: 
    try:
        loc_pm=[]
        n=nom.geocode(pm)
        loc_pm=[n.latitude,n.longitude]  
    except:
        loc_pm=[None,None]
    locs.append(loc_pm)
    
locs
```

```python
#Aqui creo los dataframes para hacer un merge y dejarlo listo para
#contnuar posteriormente con la idea de generar un archivo mkz para google Earth...
import pandas as pd

#data_transposed = zip(new_list)
#data_transposed
df1 = pd.DataFrame(new_list, columns=["Pueblo", "Estado"])
df1.head(15)
```

```python
#Creamos un segundo DataFrame...
df2=pd.DataFrame(locs,columns=["Latitud","Longitud"])
df2.head(15)
```

```python
#Inserto  una columna para hacerla mi pivote...
idx = 0
new_col = df1.Pueblo  # can be a list, a Series, an array or a scalar   
df2.insert(loc=idx, column='Pueblo', value=new_col)
df2.head(15)
```

```python
#Aqui abro una archivo cvs de Sectur que tiene el año a partir del cual los pueblos formaban parte del
#programa de Peublos Mágicos...
df_aux = pd.read_csv('Localidades_que_cuentan_con_el_nombramiento_de_Pueblos_Magicos.csv')
df3 = df_aux.Anio
df3.head(15)
```

```python
#Inserto esta columna en el dataframe 2 para hacer un solo merge...
idx = 3
new_col = df3  # can be a list, a Series, an array or a scalar   
df2.insert(loc=idx, column='Anio', value=new_col)
df2.head(15)
```

```python
#Aqui hago el merge final.
result = pd.merge(df1,
                 df2[['Pueblo','Anio','Latitud','Longitud']],
                 on='Pueblo')
result.head()
```
